Modeling

```python
x = df.drop('Target',axis=1)
y = df['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x
```

```python
x= sc.fit_transform(x)
```

```python
x
```

```python
x_trin, x_test,y_trin,y_test = train_test_split(x ,y, test_size=0.3, random_state=42)
```

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3)
from tensorflow.keras import  layers
```

```python
model = keras.Sequential([
    layers.Flatten(input_shape = (30, )),
    layers.Dense(20,activation ='relu'),
    layers.Dense(1, activation = 'sigmoid')
]

)
```

```python
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy']


)
```

```python
history =model.fit(x_trin, y_trin, epochs=100,batch_size= 32, validation_data=(x_test, y_test) )
```

```python
plt.plot(history.history['loss'], label= 'Loss')
plt.plot(history.history['val_loss'], label= 'Val Loss')
plt.title(" Loss vs Epochs")
plt.legend()# دا علشان يشغل الي قبلة
plt.show()
```

```python
plt.plot(history.history['accuracy'], label= 'accurascy')
plt.plot(history.history['val_accuracy'], label= 'val_accuracy')
plt.title ("Accuracy vs Epochs ")
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()# دا علشان يشغل الي قبلة
plt.show()
```

```python
loss,accuacy= model.evaluate(x_test, y_test)
```
